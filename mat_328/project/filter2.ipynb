```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
airbnb = pd.read_csv("nyc_airbnb_2019.csv")
airbnb.head()
```

```python
np.unique(airbnb["neighbourhood_group"])
```

__Bronx Filtering__

```python
bronx_filter = airbnb["neighbourhood_group"] == "Bronx"
bronx = airbnb[bronx_filter]

fig1 = plt.figure(figsize = (12, 5))
ax1 = fig1.add_subplot(1,2,1)
ax1.hist(bronx["price"], edgecolor = 'w', bins = 20)
ax1.set_title("Distribution of Prices Per Night in the Bronx")
ax1.set_xlabel("Prices Per Night in Dollars")
ax1.set_ylabel("Frequency")
ax2 = fig1.add_subplot(1,2,2)
ax2.hist(bronx["price"], range = (0,200), color = 'r', edgecolor = 'w', bins = 15)
ax2.set_title("Distribution of Prices Per Night in the Bronx (Post-Filter)")
ax2.set_xlabel("Prices Per Night in Dollars")
ax2.set_ylabel("Frequency")
plt.savefig("plots/bronx_price_filter_plots.png")
```

```python
fig2 = plt.figure(figsize = (14, 5))
ax3 = fig2.add_subplot(1,2,1)
ax3.hist(bronx["minimum_nights"], bins = 50)
ax3.set_title("Distribution of Minimum Nights of Listings in the Bronx")
ax3.set_xlabel("Minimum Nights")
ax3.set_ylabel("Frequency")
ax4 = fig2.add_subplot(1,2,2)
max7 = bronx["minimum_nights"] <= 31
min_nights_vc = bronx[max7]["minimum_nights"].value_counts()
ax4.bar(min_nights_vc.index, min_nights_vc.values, color = "red")
ax4.set_title("Distribution of Minimum Nights of Listings in the Bronx (Post-Filter)")
ax4.set_xlabel("Minimum Nights")
ax4.set_ylabel("Frequency")
```

__Filter__

```python
price_filter = ((airbnb["neighbourhood_group"] == "Bronx") & (airbnb["price"] <= 200)) | ((airbnb["neighbourhood_group"] == "Brooklyn") & (airbnb["price"] <= 350)) | ((airbnb["neighbourhood_group"] == "Manhattan") & (airbnb["price"] <= 500)) | ((airbnb["neighbourhood_group"] == "Queens") & (airbnb["price"] <= 300)) | ((airbnb["neighbourhood_group"] == "Staten Island") & (airbnb["price"] <= 300)) 
filtered_airbnb = airbnb[price_filter]

print(airbnb.shape)
print(filtered_airbnb.shape)
```

```python
min_nights_filter = filtered_airbnb["minimum_nights"] <= 31

filtered_airbnb = filtered_airbnb[min_nights_filter]
print(airbnb.shape)
print(filtered_airbnb.shape)
```

```python
filtered_airbnb.to_csv("filtered_nyc_airbnb_2019.csv")
```
