```python
housing["MSZoning"] = housing["MSZoning"].astype(str)
housing["Street"] = housing["Street"].astype(str)
housing["Alley"] = housing["Alley"].astype(str)
housing["LotShape"] = housing["LotShape"].astype(str)
housing["LandContour"] = housing["LandContour"].astype(str)
housing["Utilities"] = housing["Utilities"].astype(str)
housing["LotConfig"] = housing["LotConfig"].astype(str)
housing["LandSlope"] = housing["LandContour"].astype(str)
housing["Neighborhood"] = housing["Neighborhood"].astype(str)
housing["Condition1"] = housing["Condition1"].astype(str)
housing["Condition2"] = housing["Condition2"].astype(str)
housing["BldgType"] = housing["BldgType"].astype(str)
housing["HouseStyle"] = housing["HouseStyle"].astype(str)
housing["RoofStyle"] = housing["RoofStyle"].astype(str)
housing["RoofMatl"] = housing["RoofMatl"].astype(str)
housing["Exterior1st"] = housing["Exterior1st"].astype(str)
housing["Exterior2nd"] = housing["Exterior2nd"].astype(str)
housing["MasVnrType"] = housing["MasVnrType"].astype(str)
housing["ExterQual"] = housing["ExterQual"].astype(str)
housing["ExterCond"] = housing["ExterCond"].astype(str)
housing["Foundation"] = housing["Foundation"].astype(str)
housing["BsmtQual"] = housing["BsmtQual"].astype(str)
housing["BsmtCond"] = housing["BsmtCond"].astype(str)
housing["BsmtExposure"] = housing["BsmtExposure"].astype(str)
housing["BsmtFinType1"] = housing["BsmtFinType1"].astype(str)
housing["BsmtFinType2"] = housing["BsmtFinType2"].astype(str)
housing["Heating"] = housing["Heating"].astype(str)
housing["HeatingQC"] = housing["HeatingQC"].astype(str)
housing["CentralAir"] = housing["CentralAir"].astype(str)
housing["Electrical"] = housing["Electrical"].astype(str)
housing["KitchenQual"] = housing["KitchenQual"].astype(str)
housing["Functional"] = housing["Functional"].astype(str)
housing["FireplaceQu"] = housing["FireplaceQu"].astype(str)
housing["GarageType"] = housing["GarageType"].astype(str)
housing["GarageFinish"] = housing["GarageFinish"].astype(str)
housing["GarageQual"] = housing["GarageQual"].astype(str)
housing["GarageCond"] = housing["GarageCond"].astype(str)
housing["PavedDrive"] = housing["PavedDrive"].astype(str)
housing["PoolQC"] = housing["PoolQC"].astype(str)
housing["Fence"] = housing["Fence"].astype(str)
housing["MiscFeature"] = housing["MiscFeature"].astype(str)
housing["SaleType"] = housing["SaleType"].astype(str)
housing["SaleCondition"] = housing["SaleCondition"].astype(str)
```

```python
columns = housing.columns
length = len(housing)

for column in columns:
    if((np.sum(housing[column].isnull()) / length) >= 0.30):
        housing.drop([column], axis=1)
    elif ((np.sum(housing[column].isnull()) / length) <= 0.05):
        housing.dropna(axis = 0, subset = [column], inplace = True)
    else:
        mode = housing[column].mode()
        housing[column].fillna(mode, inplace=True)
```

```python
# Removing Outliers
yb_filter = (housing2["YearBuilt"] >= 1900) & (housing2["YearBuilt"] <= 2000)
housing2 = housing2[yb_filter]
bsmt_filter = housing2["TotalBsmtSF"] <= 2250
housing2 = housing2[bsmt_filter]
gr_filter = (housing2["GrLivArea"] >= 500) & (housing2["GrLivArea"] <= 3000)
housing2 = housing2[gr_filter]
sp_filter = (housing2["SalePrice"] >= 75000) & (housing2["SalePrice"] <= 400000)
housing2 = housing2[sp_filter]
```

```python
housing2.shape
```

```python
from scipy.stats import pearsonr
```

```python
plt.scatter(housing2["OverallQual"], housing2["SalePrice"], alpha = 0.05)
plt.title("Overall Quality vs. Sales Price")
plt.xlabel("Overall Quality")
plt.ylabel("Sales Price in Dollars")
```

```python
pearsonr(housing2["OverallQual"], housing2["SalePrice"])
```

__Conclusion: There is strong positive correlation between the overall quality of a house and the sales price of a house.__

```python
plt.scatter(housing2["YearBuilt"], housing2["SalePrice"], alpha = 0.25)
plt.title("Year Built vs. Sales Price")
plt.xlabel("Year Built")
plt.ylabel("Sales Price in Dollars")
```

```python
pearsonr(housing2["YearBuilt"], housing2["SalePrice"])
```

__Conclusion: There is moderate positive correlation between the year a house is built and the sales price of a house.__
