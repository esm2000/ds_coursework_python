```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
housing = pd.read_csv("train.csv")
```

```python
housing.head()
```

```python
housing2 = housing.drop(["OverallQual", "YearBuilt", "TotalBsmtSF", "GrLivArea"], axis = 1)
housing2.shape
```

```python
housing3 = housing2.select_dtypes(include =["int64", "float64"]) 
variables = housing3.columns
max_coef = 0

for variable in variables:
    
    if(variable == "SalePrice"): 
        break
    else:
        coef_tuple = pearsonr(housing2[variable], housing2["SalePrice"])
        coef = coef_tuple[0]
        if(abs(coef) > max_coef):
            max_coef = coef
            max_variable = variable
```

```python
print("The variable, " + str(max_variable) + ", has the highest coefficient of: " + str(max_coef))
```

```python
housing["total_area"] = housing["GrLivArea"] + housing["TotalBsmtSF"]
housing["area_per_room"] = housing["GrLivArea"] / housing["TotRmsAbvGrd"]
```

```python
housing_final = housing[["OverallQual", "YearBuilt", "TotalBsmtSF", "GrLivArea", "SalePrice", "GarageCars", "total_area", "area_per_room"]]
housing_final.head()
```

```python
housing_final_columns = ["OverallQual", "YearBuilt", "TotalBsmtSF", "GrLivArea", "GarageCars", "total_area", "area_per_room"]
means = []
stds = []

for column in housing_final_columns:
    mean = housing_final[column].mean()
    std = housing_final[column].std()
    
    means.append(mean)
    stds.append(std)  
```

```python
count = 0

for column in housing_final_columns:
    housing_final[column] = (housing_final[column] - means[count])
    housing_final[column] = housing_final[column] / stds[count]
    count = count + 1;
```

```python
housing_final["diff"] = np.sqrt((housing_final["OverallQual"] - -0.794879)**2 + (housing_final["YearBuilt"] - -0.339961)**2 + (housing_final["TotalBsmtSF"] - -0.399880)**2 + (housing_final["GrLivArea"] - -1.178852)**2 + (housing_final["GarageCars"] - -1.026506)**2 + (housing_final["total_area"] - -0.965146)**2 + (housing_final["area_per_room"] - -1.155675)**2)
housing_final["diff"].head()
```

```python
k = 5
index_knearest = housing_final["diff"].values.argsort()[:k]
index = housing_final.index[index_knearest]
data_knearest = housing_final.loc[index, :]
data_knearest
```

```python
prediction = data_knearest["SalePrice"].mean()
print("Using the K-Nearest Neighbor Method with k = 5, the prediction for the sale price of the first observation is: $" + str(prediction))
```
