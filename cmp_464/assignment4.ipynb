Create a new Jupyter notebook, load ratings.csv and movies.csv as dataframes.

```python
# importing pandas, numpy, pyplot, and allowing plots to show inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading in ratings and movies dataframes
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
```

1. Create column "AvgRating50" in the movies dataframe. For each movie, if there are less than 50 ratings, set its AvgRating50 value to -1.0; otherwise, set the value to the average rating.

```python
def atleast50(index):
    movie = movies.loc[index, "movieId"]
    movie_filter = ratings["movieId"] == movie
    filtered_ratings = ratings[movie_filter]
    
    if(len(filtered_ratings) >= 50):
        return filtered_ratings["rating"].mean()
    else:
        return -1.0
    
movie_indexes = movies.index
movies["AvgRating50"] = pd.Series(movie_indexes).apply(atleast50)

movies.head()
    
```

2. Print the complete list of genres. (Use split() method to split a string containing multiple genres)

```python
genres = []
def pull_genres(index):
    
    movie_genres = movies.loc[index, "genres"].split("|")
    
    for movie_genre in movie_genres:
        genres.append(movie_genre)

pd.Series(movie_indexes).apply(pull_genres)
genres = list(set(genres))

print("\n ".join(genres))
```

3. Which genre has the largest number of movies?

```python
genre_counts = []
genre_fields = movies["genres"]


for genre in genres:
    
    count = 0
    
    for i in np.arange(0,len(genre_fields)):
            
        if(genre in genre_fields[i]):
            count = count + 1
    
    genre_counts.append(count)

print(genre_counts)
```

```python
genres = pd.Series(genres)
genre_counts = pd.Series(genre_counts)

frame = {"genre": genres, "count": genre_counts}
genre_dfr = pd.DataFrame(frame)

max_filter = genre_dfr["count"] == genre_dfr["count"].max()
max_value = genre_dfr[max_filter]
max_value
```

__Conclusion:__ Drama has the highest number of movies.

4. For each genre, find the ID of the movie with highest AvgRating50 value.

```python
maxMovieIndexes = []

genre_fields = movies["genres"]
avg_ratings50 = movies["AvgRating50"]

for genre in genres:

    max_movie = 0.0
    for i in np.arange(0,len(genre_fields)):
            
        if((genre in genre_fields[i]) & (avg_ratings50[i] > max_movie)):
            max_movie = i
    
    maxMovieIndexes.append(max_movie)
```

```python
maxMovieIds = []
movieIds = movies["movieId"]

for index in maxMovieIndexes:
    maxMovieIds.append(movieIds[index])
    


maxMovieIds = pd.Series(maxMovieIds)
genre_dfr["highest_rated_id"] = maxMovieIds

genre_dfr
```

5. For each genre, find the name of the movie with the highest AvgRating50 value.

```python
maxMovieNames = []

for movie_id in maxMovieIds:
    
    movie_filter = movies["movieId"] == movie_id
    movie_name = movies[movie_filter]["title"].values[0]
    maxMovieNames.append(movie_name)
    
maxMovieNames   
    
    
```

```python
maxMovieNames = pd.Series(maxMovieNames)
genre_dfr["highest_rated_name"] = maxMovieNames

genre_dfr
```
